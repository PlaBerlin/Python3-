```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']

donation = [1.1,15.9,28.1]
x = [1,2,3]

plt.title('中国对世界经济增长的年均贡献率')

plot_x = [2.2,2.2]
y = [0,30]

plt.ylabel(u"年均贡献率")
plt.xticks(x, ["1961-1978", "1979-2012", "2013-2018"])
plt.text(1.8,17.8,'↑第2位',size='15',color = "r")
plt.text(2.8,29.8,'↑第1位',size='15',color = "r")

plt.text(0.9,1.2,'1.1%',size='15')
plt.text(1.9,16,'15.9%',size='15')
plt.text(2.9,28.2,'28.1%',size='15')


plt.bar(x=x, height=donation)
plt.plot(plot_x,y,color='r',linestyle="--",alpha=0.5)

plt.show()
```
