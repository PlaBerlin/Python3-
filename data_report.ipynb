```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.title('中国对世界经济增长的年均贡献率')

donation = [1.1,15.9,28.1]
x = [1,3,5]
plot_x = [3.2,3.2]
y = [0,35]

plt.ylabel(u"年均贡献率")
plt.xticks(x, ["1961-1978", "1979-2012", "2013-2018"])

plt.text(2.5,10,r'$\uparrow$第2位',size='15',color = "r")
plt.text(4.5,20,r'$\uparrow$第1位',size='15',color = "r")
plt.text(0.8,1.2,'1.1%',size='15')
plt.text(2.7,16,'15.9%',size='15')
plt.text(4.7,28.2,'28.1%',size='15')

plt.bar(x=x, height=donation,alpha=0.7)
plt.plot(plot_x,y,color='r',linestyle="--",alpha=0.5)
plt.text(2.9,25,'2006年',size='15')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']

times = [1962, 1978, 2018]
GNI= [70, 200, 9470]
globel = [70/0.146, 200/0.1, 9470/0.853]
times2 = [1962, 1978,2018]
plot_x = [1978,1978]
plot_y = [0,10000]

plt.scatter(times,GNI,label='中国',marker='x')
plt.scatter(times2,globel, label='世界平均水平',marker='x')
# plt.plot(plot_x,plot_y,color='r',alpha=0.5)

plt.title('人均国民总收入')
plt.ylabel('单位:美元')
plt.text(1978,200,'175位',size='15',color = "r")
plt.text(2018,9470,'71位',size='15',color = "r")

plt.annotate('改革开放', xy=(1978, 200), xytext=(1965, 1000),
            arrowprops=dict(facecolor='red', shrink=0.01),
            )

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.title('我国GDP占世界总量')

gdp = np.array([1.8,11.4,15.9])
x =  np.array([1978,2012,2018])


plt.ylim((0, 17.5))
plt.ylabel(u"占比")

plt.scatter(x,gdp,marker='x')

for x,y in zip(x,gdp):
    plt.text(x,y+0.5,'%.2f' %y+'%', ha='center',size=12)
    

plt.annotate('第11位', xy=(1978, 1.8), xytext=(1982, 3),
            arrowprops=dict(facecolor='red', shrink=0.01),size=15
            )
plt.annotate('超过日本', xy=(2012, 11.4), xytext=(2000, 10),
            arrowprops=dict(facecolor='red', shrink=0.01),size=15
            )

plt.annotate('世界第二', xy=(2018, 15.9), xytext=(2006, 14),
            arrowprops=dict(facecolor='red', shrink=0.01),size=15
            )
plt.show()
```
